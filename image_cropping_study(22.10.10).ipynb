## Ï†ÑÏ≤¥ ÏΩîÎìú
```python
import cv2
import numpy as np

img = cv2.imread("313-2-1-15-Z99_A-0000001.jpg")
rows, cols, _ = img.shape
#Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï∂úÎ†•
print("Rows", rows)
print("Cols", cols)

#Ïù¥ÎØ∏ÏßÄ ÏûêÎ•¥Í∏∞
cut_image=img[200:950, 800:1200]


cv2.imshow("Cut Image",cut_image)

#ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ
cv2.imshow("image", img)

#Ï∂ïÏÜå
scaleX = 0.6
scaleY = 0.6
#resizing ÏΩîÎìú
scaledown_img = cv2.resize(cut_image, None, fx=scaleX, fy=scaleY, interpolation = cv2.INTER_LINEAR)
cv2.imshow("Scale Down", scaledown_img)

#scaledown Ìïú img Ï†ÄÏû•
save_file = 'img_crop_scaledown.jpg' #Ï†ÄÏû•Ìï† Ïù¥Î¶Ñ
cv2.imwrite(save_file, scaledown_img) #ÌååÏùºÎ°ú Ï†ÄÏû•

cv2.waitKey(0) #ÏûÖÎ†• ÏûàÏùÑ ÎïåÍπåÏßÄ ÎåÄÍ∏∞

```

#### üíªImage Cropping ÏΩîÎìú
```python
import cv2
import numpy as np

img = cv2.imread("313-2-1-15-Z99_A-0000001.jpg")
rows, cols, _ = img.shape
#Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï∂úÎ†•
print("Rows", rows)
print("Cols", cols)

#Ïù¥ÎØ∏ÏßÄ ÏûêÎ•¥Í∏∞
cut_image=img[200:950, 800:1200]
#200:950, 800:1200 = ÏÉÅ:Ìïò, Ï¢å:Ïö∞

cv2.imshow("Cut Image",cut_image)

#ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ
#cv2.imshow("image", img)

cv2.waitKey(0) #ÏûÖÎ†• ÏûàÏùÑ ÎïåÍπåÏßÄ ÎåÄÍ∏∞

```
<br>

#### üíªÏ†ÄÏû• ÏΩîÎìú

```python
#Ï†ÄÏû•
save_file = 'img_crop.jpg' #Ï†ÄÏû•Ìï† Ïù¥Î¶Ñ
cv2.imwrite(save_file, cut_image) #ÌååÏùºÎ°ú Ï†ÄÏû•
```
<br>

#### üíªÏù¥ÎØ∏ÏßÄ Ï∂ïÏÜå ÏΩîÎìú

```python
#resizing ÏΩîÎìú
scaledown_img = cv2.resize(cut_image, None, fx=scaleX, fy=scaleY, interpolation = cv2.INTER_LINEAR)
cv2.imshow("Scale Down", scaledown_img)

#scaledown Ìïú img Ï†ÄÏû•
save_file = 'img_crop_scaledown.jpg' #Ï†ÄÏû•Ìï† Ïù¥Î¶Ñ
cv2.imwrite(save_file, scaledown_img) #ÌååÏùºÎ°ú Ï†ÄÏû•

#Ï†ÄÏû•
save_file = 'img_crop_scaledown.jpg' #Ï†ÄÏû•Ìï† Ïù¥Î¶Ñ
cv2.imwrite(save_file, scaledown_img) #ÌååÏùºÎ°ú Ï†ÄÏû•

```
